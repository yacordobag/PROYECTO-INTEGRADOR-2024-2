```python
import pandas as pd

file_path = '/content/drive/MyDrive/medellin_clima_2_anios.csv'

data = pd.read_csv(file_path)

# Clasificación de columnas
temp_columns = [col for col in data.columns if col.startswith('t_')]
nivel_columns = [col for col in data.columns if col.startswith('nivel_metros_')]
pluvio_columns = [col for col in data.columns if col.startswith('pr_') or col.startswith('p_')]

# Calcular estadísticas básicas
temp_stats = data[temp_columns].describe()
nivel_stats = data[nivel_columns].describe()
pluvio_stats = data[pluvio_columns].describe()

# Mostrar resultados
print("Estadísticas de temperatura:")
print(temp_stats)

print("\nEstadísticas de nivel:")
print(nivel_stats)

print("\nEstadísticas pluviométricas:")
print(pluvio_stats)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

file_path = '/content/drive/MyDrive/medellin_clima_2_anios.csv'

data = pd.read_csv(file_path)

# Filtrar las columnas que contienen 'promedio' en su nombre y no contienen 'vv'
promedio_columns = [col for col in data.columns if 'promedio' in col and 'vv' not in col]

# Calcular la matriz de correlación solo para las columnas "promedio" que no incluyen "vv"
correlation_matrix = data[promedio_columns].corr()

# Mostrar la matriz de correlación
print("Matriz de correlación entre las variables promedio:")
print(correlation_matrix)

# Crear un mapa de calor para la matriz de correlación
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)
plt.title("Matriz de Correlación - Variables Promedio")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

file_path = '/content/drive/MyDrive/medellin_clima_2_anios.csv'

data = pd.read_csv(file_path)

# Filtrar las columnas que contienen 'promedio' en su nombre y no contienen 'vv'
promedio_columns = [col for col in data.columns if 'promedio' in col and 'vv' not in col]

# Crear un histograma para cada columna "promedio" (sin "vv")
plt.figure(figsize=(12, 8))
for i, col in enumerate(promedio_columns, 1):
    plt.subplot(2, 3, i)  # 2 filas, 3 columnas para los subgráficos
    sns.histplot(data[col], kde=True)  # Histograma con curva de densidad
    plt.title(f'Histograma de {col}')
    plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

file_path = '/content/drive/MyDrive/medellin_clima_2_anios.csv'

data = pd.read_csv(file_path)

# Filtrar las columnas que contienen 'promedio' en su nombre y no contienen 'vv'
promedio_columns = [col for col in data.columns if 'promedio' in col and 'vv' not in col]

# Crear un gráfico de dispersión entre todas las variables "promedio" (sin "vv")
plt.figure(figsize=(12, 8))
sns.pairplot(data[promedio_columns])
plt.suptitle("Gráficos de Dispersión entre Variables Promedio", y=1.02)
plt.show()
```
