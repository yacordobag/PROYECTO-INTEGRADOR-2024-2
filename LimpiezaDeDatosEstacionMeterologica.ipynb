# Precision

```python
def detectar_outliers_iqr(df, columna):
    Q1 = df[columna].quantile(0.25)
    Q3 = df[columna].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[columna] < (Q1 - 1.5 * IQR)) | (df[columna] > (Q3 + 1.5 * IQR))]
    return outliers

for columna in ['h', 't', 'pr', 'vv', 'vv_max', 'dv', 'dv_max', 'p']:
    outliers = detectar_outliers_iqr(df, columna)
    print(f"Outliers en la columna {columna}:\n", outliers[['codigo', 'fecha_hora', columna]])
```

# Completitud

```python
completitud = pd.DataFrame({
    'Total Valores': df.shape[0],
    'Valores Completos': df.notnull().sum(),
    'Valores Faltantes': df.isnull().sum(),
    'Porcentaje Completitud (%)': (df.notnull().mean() * 100).round(4)
})
completitud
```

```python
df_limpieza = df.dropna(subset=['fecha_hora'])
df_limpieza.reset_index(drop=True, inplace=True)
```

```python
columnas_con_datos = ['h', 't', 'pr', 'vv', 'vv_max', 'dv', 'dv_max', 'p']
```

```python
def procesar_grupo(grupo):
    grupo = grupo.sort_values(by='fecha_hora')
    return grupo

df_limpieza = (
    df_limpieza.groupby('codigo', group_keys=False)
    .apply(procesar_grupo)
    .reset_index(drop=True)
)

df_limpieza
```

```python
df_limpieza[columnas_con_datos] = (
    df_limpieza.groupby('codigo', group_keys=False)[columnas_con_datos]
    .apply(lambda group: group.interpolate(method='linear'))
    .reset_index(drop=True)
)
df_limpieza
```

```python
completitud = pd.DataFrame({
    'Total Valores': df_limpieza.shape[0],
    'Valores Completos': df_limpieza.notnull().sum(),
    'Valores Faltantes': df_limpieza.isnull().sum(),
    'Porcentaje Completitud (%)': (df_limpieza.notnull().mean() * 100).round(4)
})
completitud
```

```python
# Mirar completitud de la fecha 
df = df.sort_values(by=['codigo', 'fecha_hora']).reset_index(drop=True)
df['diferencia_tiempo'] = df.groupby('codigo')['fecha_hora'].diff().dt.total_seconds() / 60 
saltos_minuto = df[(df['diferencia_tiempo'] != 1) & (df['diferencia_tiempo'].notnull())]
print("Registros que no están en intervalos de un minuto:")
saltos_minuto[['codigo', 'fecha_hora', 'diferencia_tiempo']]
```

```python
saltos_minuto_ordenado = saltos_minuto[['codigo', 'fecha_hora', 'diferencia_tiempo']].sort_values(by='diferencia_tiempo', ascending=False)
saltos_minuto_ordenado.head(len(df["codigo"].unique()))
```

```python
saltos_mayores_por_codigo = saltos_minuto.sort_values(by=['codigo', 'diferencia_tiempo'], ascending=[True, False]) \
                                         .drop_duplicates(subset=['codigo'], keep='first') \
                                         [['codigo', 'fecha_hora', 'diferencia_tiempo']]
saltos_mayores_por_codigo
```

```python
# Agrupacion por hora
df['fecha_hora'] = pd.to_datetime(df['fecha_hora'])
df['fecha_hora_hora'] = df['fecha_hora'].dt.floor('h')
resultados_por_hora = df.groupby(['codigo', 'fecha_hora_hora']).agg({
    'h': ['mean', 'max', 'min'],
    't': ['mean', 'max', 'min'],
    'pr': ['mean', 'max', 'min'],
    'vv': ['mean', 'max', 'min'],
    'vv_max': ['mean', 'max', 'min'],
    'dv': ['mean', 'max', 'min'],
    'dv_max': ['mean', 'max', 'min'],
    'p': ['mean', 'max', 'min'],
    'calidad': lambda x: x.mode()[0] 
})
```

```python
resultados_por_hora.columns = ['_'.join(col).strip() for col in resultados_por_hora.columns.values]
resultados_por_hora = resultados_por_hora.reset_index()
resultados_por_hora
```
